```python
import requests
import requests_cache
requests_cache.install_cache('demo_cache')

import pandas as pd

#Getting Library Information
r = requests.get("https://proteomics3.ucsd.edu/ProteoSAFe/LibraryServlet?library=GNPS-MSMLS")
spectra_list = r.json()["spectra"]

spectra_df = pd.DataFrame(spectra_list)
```

```python
spectra_df.keys()
```

```python
#Loading the input to MS/MS Chooser

input_urls = ["ftp://massive.ucsd.edu/MSV000084496/other/GNPS00007%20MSMS-Chooser%20Template%20-%20MSMS-Chooser%20Submission.tsv", 
              "ftp://massive.ucsd.edu/MSV000084495/other/GNPS00006%20MSMS-Chooser%20Template%20-%20MSMS-Chooser%20Submission.tsv",
              "ftp://massive.ucsd.edu/MSV000084479/updates/2019-10-21_kcweldon_012df770/other/GNPS00005_Template_Correct.tsv",
              "ftp://massive.ucsd.edu/MSV000084493/other/GNPS00004%20MSMS-Chooser%20Template%20-%20MSMS-Chooser%20Submission.tsv",
              "ftp://massive.ucsd.edu/MSV000084492/other/GNPS00003%20MSMS-Chooser%20Template%20-%20MSMS-Chooser%20Submission.tsv",
              "ftp://massive.ucsd.edu/MSV000084494/other/GNPS00002%20MSMS-Chooser%20Template%20-%20MSMS-Chooser%20Submission.tsv",
              "ftp://massive.ucsd.edu/MSV000084072/updates/2019-09-26_fevargas_7e0e47c2/other/Test%20for%20Chemical%20Standard%20to%20GNPS%20Library%20Spectrum_test1.txt"
             ]

for input_url in input_urls:
    pd.DataFrame()
    
msms_chooser_df = pd.concat([pd.read_csv(url, sep="\t") for url in input_urls], ignore_index=True)
grouped_structure = msms_chooser_df.groupby(["SMILES"]).first()
grouped_structure["SMILES"] = grouped_structure.index
grouped_structure.head()
```

```python
# Summarizing
import urllib.parse
output_records = []

for record in grouped_structure.to_dict(orient="records"):
    smiles = record["SMILES"]
    
    output_dict = {}
    output_dict["SMILES"] = smiles
    output_dict["COMPOUND_NAME"] = record["COMPOUND_NAME"]
    
    filtered_df = spectra_df[spectra_df["Smiles"] == smiles]    
    output_dict["ms2_count"] = len(filtered_df)
    output_dict["adducts"] = ",".join(list(filtered_df["Adduct"]))
    
    #Getting mass
    try:
        r = requests.get("https://gnps-structure.ucsd.edu/structuremass?smiles={}".format(urllib.parse.quote(smiles)))
        if r.status_code == 200:
            output_dict["mass"] = float(r.text)
        else:
            continue
    except:
        continue
    
    output_records.append(output_dict)
    
output_df = pd.DataFrame(output_records)
output_df.to_csv("summary.tsv", sep="\t", index=False)
```
